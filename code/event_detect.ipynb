```python
# Dividing the data into windows of size 100
windows = []

last = len(data) // 100

for i in range(last):
    windows.append(data.iloc[i*100:i*100+100])
    
windows.append(data.iloc[last*100:-1])
```

```python
j=1
for i in windows:
    local_max = i['z1'].max()
    local_min = i['z1'].min()
    if((max_threshold-1)<local_max<=max_threshold  and (min_threshold+1)<local_min<=min_threshold):
        i.to_csv('/home/surbhi/deadline_project/data_26_nov_2019/driver type/driver2/event_type/speed_breaker/TWO_WHEELER_MOUNTER_26_11_2019-06_32_10_ONEPLUS A3003_150/event_detect/event'+str(j)+'.csv')
        j=j+1
```
